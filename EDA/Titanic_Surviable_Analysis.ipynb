```python
# Feature Engineering
```

```python
df['Family size'] = df['Parch'] + df['SibSp']
```

```python
df
```

```python
def Family_type(number):
        if(number == 0):
            return 'Alone'
        elif(number > 0 and number < 4):
            return 'Medium'
        else:
            return 'Large'
```

```python
df['Family_type'] = df['Family size'].apply(Family_type)
```
